```python
df = pd.read_csv('datas/korea-star.csv')
df
```

```python
# 먼저 apply에 활용할 함수를 정의합니다.
def gender(x):
    if x == '남자':
        return 1
    elif x == '여자':
        return 0
    else:
        return -1
```

```python
# python map() 함수 사용
def a_func(x):
    return x * 2

a = [1, 2, 3, 4]

result = map(a_func, a)
# for test in result:
#     print(test) : map 도 덩어리 : 반복문 사용
print(result)
print(type(result))
print(list(result))
```

```python
# df['성별'] = df['성별'].apply(gender)
df['성별_NEW'] = df['성별'].apply(gender)
```

## [문제] apply 함수 활용 및 컬럼 추가
- (목표) cm당 인기지수 구하기
- (인기지수/ 키)를 구하여 '인기지수/키' 컬럼 추가

ex
- 키: 174, 인기지수: 100000
- 계산 식: 100000 / 174

```python
df = pd.read_csv('datas/korea-star.csv')
```

### 1. apply에 적용할할 함수 정의

```python
def cm_to_brand1(x):
    # print(x["키"])
    result = x["인기지수"]/x["키"]
    # print(result)
    return round(result, 2)
```

```python
def cm_to_brand(x):
    result = x["인기지수"] / x["키"]
    return result
```

### 2. df.apply(호출함수) 실행

```python
df["키/인기지수"] = df[["인기지수", "키"]].apply(cm_to_brand1, axis=1)
```

```python
df["키/인기지수"] = df.apply(cm_to_brand, axis=1)
```

```python
df
```

## 1-3. lambda 함수의 적용
- lambda 함수 : 한줄 함수
- 식별자(변수) = lambda 매개변수1, 매개변수2 : 수행문
- f = lambda x, y : x + y

```python
# 실행결과 식별 = f(5, 6)

f = lambda x, y : x+y
f(5, 6)
```

```python
def lambda_ex(num1, num2):
    return num1 * num2
lambda_ex(f(6, ), 2)
```

```python
def lambda_ex(num1, num2):
    return num1 * num2

test = [1]
test1 = [2] 

lambda_ex(int(list(map(lambda x, y : x+y, test, test1))[0]), 2)
```

# 3. map() - 값을 매핑!

```python
df = pd.read_csv('datas/korea-star.csv')
df.head()
```

#### df.map(mapping) 상용 방법
- mapping: dictionary
<pre>
{
    원래값1:변경값1,
    원래값2:변경값2,
    ...
    원래값n:변경값n
}
</pre>

```python
print(df["성별"].value_counts()) # 정보 확인
print(df["성별"].unique())
```

```python
mapping = {
    "여자" : 0,
    "남자" : 1
}
```

```python
df["성별"] = df["성별"].map(mapping)
```

```python
df.head()
```

```python
df.info()
```

```python
# mapping 원래값 type을 맞춰줘야 함
mapping = {
    "0": "female",
    "1": "male"
}
```

```python
df["성별"] = df["성별"].map(mapping)
df.head()
```
