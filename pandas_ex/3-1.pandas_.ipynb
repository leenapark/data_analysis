## 1. 결측값을 채워주는 fillna
- **fillna()**: nan 값에 대하여 지정한 값을 채워주는 함수

```python
# 결측치 확인
df.info()
```

'키'에 2개의 데이터가 누락, 그룹에 2개의 데이터가 누락된 것 확인

키가 없는 사람은 존재할 수 없기 때문에,<br>
데이터가 잘 못 되어 있다는 것을 표기 하기 위하여,<br>
**누락된 데이터를 -1로 채워봅시다.**

```python
df['키']
```

```python
df['키'].fillna(-1)
```

키의 NaN 값을 채워준다음 데이터 프레임에 적용하기 위한 방법은

- inplace=True 옵션을 주거나,
- fillna로 채워 준 값을 다시 대입
- **fillna로 채워준 값을 다시 대입 해 주는 방식을 추천 함**

```python
df2 = df.copy()
```

```python
df2
```

```python
df2['키'] = df2['키'].fillna(-1)
# df2['키'].fillna(-1, inplace=True)
```

```python
df2["키"]
```

기술통계 값으로 채우기

- 최빈값 mode()
- 중앙값 median()
- 평균값 mean()

```python
# [문제] 키의 최빈값 평균 계산
print(df["키"].mode())
print(df["키"].mode().mean())
```

```python
# [문제] 키 컬럼의 중앙값 계산

print(df["키"].median())
```

```python
# [문제] 키 컬럼의 평균을 소숫점 1에서 반올림하기

print(df["키"].mean().round(0))
```

```python
# [문제] 
# - 키 컬럼에 키의 평균을 내서 소숫점 1에서 반올림한 값으로 계산
# - 계산한 값을 NaN 값 대신 채워 넣기

print(df["키"].fillna(df["키"].mean().round(0)))
```

## 2. 빈 값(NaN)이 있는 행을 제거
df.dropna() 사용

```python
df.info()
```

```python
df
```

```python
df.dropna()
```

#### **dropna()는 몇가지 옵션**

- axis
    - 행 0
    - 열 1
- 'how = any': 한개라도 있는 경우 드랍
- 'how = all' : 모두 NaN인 경우 드랍

### 2-1. axis nan 값이 있는 행 / 열 삭제
axis=0 : 데이터 프레임에 nan 값이 있는 행 삭제

```python
df2 = df.copy()
df2
```

```python
df2.dropna(axis=0)
```

##### axis=1 : 데이터프레임에 nan 값이 있는 열 삭제

```python
df2.dropna(axis=1)
```

### 2-2. how= 옵션 패러미터
- 'any': 한개라도 있는 경우 드랍
- 'all' : 모두 NaN인 경우 드랍

```python
df2
```

```python
df2.dropna(axis=0, how='any')
```

```python
df2.dropna(axis=0, how='all')
```

```python
import numpy as np
```

```python
# index 3 에 nan으로 채워넣기
# df2.iloc[3] = np.nan
df2.loc[len(df2)] = np.nan
```

```python
df2
```

```python
#  행 모두가 nan 이면 모두 삭제
df2.dropna(axis=0, how='all')
```

##### keep 옵션으로 유지하고 싶은 데이터를 선택 가능
- keep: 'first' / 'last'
    - first :첫번째것 유지

```python
df = pd.read_csv("datas/korea-star.csv")
df['키']
```

```python
df['키'].drop_duplicates(keep='last')
```

```python
df['키'] = df['키'].drop_duplicates(keep='last')
```

```python
df
```

## 4. Drop - column/row 제거하기

```python
df = pd.read_csv("datas/korea-star.csv")
```

```python
df.head()
```

### 4-1. drop() - column 제거하기
drop()을 활용한 column을 삭제
- axis=1 옵션

```python
df2 = df.copy()
print(df2)
```

```python
df2 = df2.drop('그룹', axis=1)
print(df2)
```

#### [문제] : '그룹','소속사', '생년월일' 컬럼 삭제
- 복수의 column을 삭제 하고 할 때는 list로 지정

```python
df2 = df2.drop(["소속사", "생년월일"], axis=1)
print(df2)
```

### 4-2. drop() - row 제거하기
row를 제거할 때 : 제거를 원하는 index와 axis=0 옵션 사용

```python
df2 = df2.drop(3, axis=0)
print(df2)
```

#### [문제] row 3, 5 삭제
- 복수의 row를 제거하고자 할 때는 list로 지정함

```python
df2 = df2.drop([3, 5], axis=0)
print(df2)
```

## 실습
[실습 사이트](https://wikidocs.net/154787)

```python
idx = ['row1','row2','row3','row4','row5']
col = ['col1','col2','col3']
data= [['A','가',1],['A','가',1],['A','나',2],['B','나',3],['B','다',4]]
df = pd.DataFrame(data, idx, col)
print(df)
```

```python
print(df.duplicated(keep='first'))
# row1    False
# row2     True # row1과 row2는 모든 요소가 같고, 위에서부터 확인시 row2가 중복값이 됩니다. -> True반환 
# row3    False
# row4    False
# row5    False
```

```python
print(df.duplicated(subset=['col1','col3']))
```

### subset
- subset을 이용하여 특정 열을 대상으로만 중복행의 확인이 가능합니다.

```python
print(df.duplicated(subset=['col1','col3']))
print(df.duplicated(subset=['col2', 'col1']))
print(df.duplicated(subset=['col1']))
print(df.duplicated(subset=['col2']))
print(df.duplicated(subset=['col3']))
```

## drop duplicated

```python
print(df.drop_duplicates(subset=['col1','col3']))
print(df.drop_duplicates(subset=['col2', 'col1']))
print(df.drop_duplicates(subset=['col1']))
print(df.drop_duplicates(subset=['col2']))
print(df.drop_duplicates(subset=['col3']))
```
