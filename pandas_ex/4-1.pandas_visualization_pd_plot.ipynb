## 1. Plot 그래프
plot은 일반 선그래프를 나타냅니다.

```python
df = pd.read_csv('datas/house_price_clean.csv')
```

kind 옵션을 통해 원하는 그래프를 그릴 수 있습니다.

kind 옵션:

- line: 선그래프
- bar: 바 그래프
- barh: 수평 바 그래프
- hist: 히스토그램
- kde: 커널 밀도 그래프
- hexbin: 고밀도 산점도 그래프
- box: 박스 플롯
- area: 면적 그래프
- pie: 파이 그래프
- scatter: 산점도 그래프

데이터의 성격에 맞는 그래프 형식이 있습니다. 앞으로 차근차근 살펴보도록 하겠습니다.

## line 그래프
- line 그래프는 데이터가 연속적인 경우 사용하기 적절합니다. (예를 들면, 주가 데이터)


```python
df['분양가'].plot(kind='line')
```

```python
df_seoul = df.loc[df['지역'] == '서울']
df_seoul
```

```python
df_seoul.groupby('연도')['분양가'].mean()
```

```python
# 년도별 평균 분양가 추이 시각화
df_seoul.groupby('연도')['분양가'].mean().plot(kind='line')
```

## bar 그래프
bar 그래프는 그룹별로 비교할 때 유용합니다.

```python
df.groupby('지역')['분양가'].mean()
```

```python
df.groupby('지역')['분양가'].mean().plot(kind='bar')
```

```python
df.groupby('지역')['분양가'].mean().plot(kind='barh')
```

## 히스토그램 (hist)
- 참고 :https://wikidocs.net/160069

히스토그램은 **분포-빈도 를 시각화**하여 보여줍니다

가로축에는 분포를, 세로축에는 빈도가 시각화되어 보여집니다.

```python
df['분양가'].plot.hist(bins=10)
```

## 커널 밀도 그래프

- 참고 : https://wikidocs.net/160072
- scipy 설치 해야함
    - pip install scipy
- 히스토그램과 유사하게 밀도를 보여주는 그래프
- 히스토그램과 유사한 모양새를 갖추고 있음
- 부드러운 라인을 가지고 있음

```python
df['분양가'].plot.kde()
```

## Hexbin
- 참고 : https://wikidocs.net/160049
- hexbin은 고밀도 산점도 그래프임
- x와 y 키 값을 넣어 주어야함
- x, y 값 모두 numeric 한 값을 넣어 주어야함.
- 데이터의 밀도를 추정함

```python
df.plot(kind='hexbin', x='분양가', y='연도', gridsize=20)
# gridsize=20 조정해 보기
```

## 박스 플롯(box)
- 참고 : https://wikidocs.net/160069

```python
df_seoul = df.loc[df['지역'] == '서울']
```

```python
df_seoul['분양가'].plot.box()
plt.show()
```

```python
from IPython.display import Image

Image('https://cdn1.byjus.com/wp-content/uploads/2020/10/Box-Plot-and-Whisker-Plot-1.png')
```

```python
df_seoul.describe()
```

IQR은 Inter Quantile Range의 약어로써, (3Q - 1Q) * 1.5 값입니다.

```python
IQR = (7732 - 6519.75) * 1.5
IQR
```

```python
## 박스 플롯 max
7732 + IQR
```

```python
## 박스 플롯 min
6519.75 - IQR
```

박스 플롯은 데이터 outlier 감지 할 때 가장 많이 활용 되며, 25%, median, 75% 분위의 값을 활용하는 용도로 많이 활용됩니다.

오류나 극단적 데이터를 배제하기 위해서 bot플롯 사용

## area plot
- 참고 : https://wikidocs.net/160034
- area plot은 line 그래프에서 아래 area를 모두 색칠해 주는 것이 특징입니다.

```python
df.groupby('월')['분양가'].count().plot(kind='line')
# df.groupby('월')['분양가'].count().plot.line()
```

```python
# df.groupby('월')['분양가'].count().plot(kind='area')
df.groupby('월')['분양가'].count().plot.area()
```

## pie plot (파이 그래프)
pie는 대표적으로 데이터의 점유율을 보여줄 때 유용합니다.

```python
# df.groupby('연도')['분양가'].count().plot(kind='pie')
df.groupby('연도')['분양가'].count().plot.pie()
```

## scatter plot (산점도 그래프)
- 참고 : https://wikidocs.net/160049
- 점으로 데이터를 표기해 줍니다
- x, y 값을 넣어주어야합니다 (hexbin과 유사)
- x축과 y축을 지정해주면 그에 맞는 데이터 분포도를 볼 수 있습니다.
- 역시 numeric 한 column 만 지정할 수 있습니다

```python
df.plot(x='월', y='분양가', kind='scatter', colormap="hot")
# df.plot().scatter(x='월', y='분양가', s=3, c='c', colormap="hot")
plt.show()
```

```python
df.plot(x='연도', y='분양가', kind='scatter')
```

```python
import numpy as np
df_num = pd.DataFrame({'x':np.random.randn(10000),
                   'y':np.random.randn(10000),
                   'c':np.random.randint(0,10,size=10000)})
```

```python
df_num.plot.hexbin(x='x',y='y',gridsize=20,colormap='viridis',C='c')
plt.show()
```

```python
df_num.plot.scatter(x='x',y='y',s=3,c='c',colormap='hot')
plt.show()
```
