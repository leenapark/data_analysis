## 2. DataFrame 병합하기 (merge)
- concat() : row나 column 기준으로 단순하게 이어 붙히기 였다면(index를 기준으로),
- merge() : 특정 고유한 키(unique id) 값을 기준으로 병합하기

```python
df1 = pd.read_csv('datas/korea-star.csv')
df2 = pd.read_csv('datas/korea-star-pay.csv')
```

```python
df1
```

```python
df2
```

df1와 df2는 **'이름' column**이 겹친다.

'이름'을 기준으로 두 DataFrame을 **병합**할 수 있음

**pd.merge(left, right, on = '기준column', how = 'left')**
- left : 왼쪽에 위치할 데이터프레임
- right : 오른쪽에 위치할 데이터프레임
- on : 병합의 기준이 되는 column 설정
- how : 'left', 'right', 'inner', 'outer' 라는 4가지의 병합 방식중 한가지 선택

### 2-1. how= 옵션의 left, right 방식
#### merge 수행을 위한 데이터 준비
- df2 에서 5개의 데이터를 고의적으로 drop 함.

```python
df_right = df2.drop([1, 3, 5, 7, 9], axis=0)
df_right
```

```python
df_right = df_right.reset_index(drop=True)
```

```python
df_right
```

병합할 2개의 DataFrame의 행의 갯수가 맞지 않음.(실제 충분히 있을 수 있는 사례)

'how=left' 파라미터 옵션 사용 : left 데이터 프레임으로 병합
- left DataFrame에 키 값이 존재하면 해당 데이터를 유지하고
- 병합한 right DataFrame에 NaN 값 대입

```python
df_result = pd.merge(df1, df_right, on="이름", how="left")
df_result
```

**'how=right' 옵션 : right DataFrame을 기준으로 병합**

left DataFrame이 더 많은 데이터를 보유 있으면, right의 기준컴럼과 비교하여 없는 데이터는 삭제 됨  
DataFrame 사이즈가 줄어듬

```python
df_result2 = pd.merge(df1, df_right, on="이름", how="right")
df_result2
```

### 2-2. how옵션의 inner, outer 방식
- inner 방식(교집합) : 두 DataFrame에 키(on) 값이 겹치는 경우만 병합.
- outer 방식 (합집합) : 두 DataFrame에 키(on) 값을 기준으로 모두 병합. 없는 값은 NaN으로 대입됨

```python
df1
```

```python
df_right
```

```python
df_left = df1.drop([2, 4, 8], axis=0)
df_left.reset_index(drop=True)
```

```python
df_inner = pd.merge(df_left, df_right, on="이름", how="inner") # 양 데이터의 교집합 데이터 이외 전부 삭제
df_inner
```

```python
df_outer = pd.merge(df_left, df_right, on="이름", how="outer") # 모든 데이터 합쳐짐
df_outer
```

## 2-3. column명은 다르지만, 동일한 성질의 데이터 인 경우?
예를 들면, 이름과 성함

```python
df1
```

```python
df_right
```

```python
df_right.columns = ['성함', '연봉', '가족수']
df_right
```

df1와 df_right를 병합하려고 했더니,  
df1에서는 '이름', df_right에서는 '성함'으로 표기되어  
같은 column이 존재하지 않으므로  
**기준이 되는 column**을 지정할 수 없음.

두 DataFrame에 대해서 모두 설정 : left_on, right_on 파라미터 옵션을 줌

```python
df_rename = pd.merge(df1, df_right, left_on='이름', right_on='성함', how='outer')
df_rename
```

## [문제] '성함'컬럼 삭제 하기

```python
df_drop = df_rename.drop("성함", axis=1)
df_drop
```
