# matplotlib 데이터 시각화 2
## 1. 데이터 시각화란?
### 정의 :
- 데이터를 시각적으로 나타내는 과정
- 데이터를 그래프나 차트를 사용하여 표현하는 것
### 목표:
- 데이터 분석과 이해를 잘 하기 위함
- 데이터의 상태를 직관적으로 이해하고, 인사이트를 도출하기 위함
### 목적:
- 비교, 트렌드 파악, 전체 대비 부분 파악, 상관관계, 연결관계, 공간 정보 파악 등 파악하기 위함
### 종류:
- 시각화를 위한 오픈 소스 lib : matplotlib, seaborn, plotly, ggplot 등

## 2. 데이터 시각화 유형
- 데이터 차트 유형은 다양한 목적에 따라 선택해서 사용할 수 있음
- 비교, 트렌드 파악, 전체 대비 부분 파악, 상관관계, 연결관계, 공간 정보 파악 등

### 목적에 따른 분류
#### 1. 비교(Comparison):
- 데이터의 여러 항목을 비교하고자 할 때 유용한 차트 유형
- 막대 그래프(Bar Plot), 라인 그래프(Line Plot), 박스 플롯(Box Plot) 등을 사용하여 데이터 간 비교를 시각화함
- 예시: 다른 도시의 인구 비교를 막대 그래프로 표현하여 각 도시의 인구 차이 비교

#### 트렌드 파악(Trend Analysis):
- 데이터의 추이와 변화를 파악하고자 할 때 사용되는 차트 유형
- 라인 그래프(Line Plot), 영역 그래프(Area Plot), 히스토그램(Histogram) 등을 사용하여 데이터의 변화를 시각화함
- 예시: 월별 매출액을 라인 그래프로 표현하여 매출 트렌드 파악

#### 전체 대비 부분 파악(Comparison to Whole):
- 데이터의 부분이 전체에 대비하여 어떤 비율을 차지하는지 파악하고자 할 때 사용되는 차트 유형
- 원 그래프(Pie Chart), 도넛 그래프(Donut Chart) 등을 사용하여 데이터의 상대적인 비율을 시각화함
- 예시: 매출액 중 각 제품군의 비율을 원 그래프로 표현하여 각 제품군이 전체 매출액에서 차지하는 비율 파악

#### 상관관계(Correlation):
- 두 변수 간의 관계와 상관관계를 파악하고자 할 때 사용되는 차트 유형
- 산점도(Scatter Plot), 히트맵(Heatmap) 등을 사용하여 변수 간의 상관관계를 시각화함
- 예시: 공부 시간과 시험 점수를 산점도로 표현하여 두 변수 간의 상관관계를 파악함

#### 연결관계(Connection):
- 데이터 간의 연결관계를 시각화하고자 할 때 사용되는 차트 유형
- 네트워크 그래프(Network Graph), 트리맵(Treemap) 등을 사용하여 데이터의 연결관계를 시각화함
- 예시: SNS에서 사용자 간의 친구 관계를 네트워크 그래프로 표현하여 사용자들 간의 연결관계 파악

#### 공간 정보 파악(Spatial Analysis):
- 지리적인 위치나 공간 데이터에 대한 정보를 파악하고자 할 때 사용되는 차트 유형
- 지도(Map), 히트맵(Heatmap) 등을 사용하여 데이터의 공간적 분포를 시각화함
- 예시: 도시의 인구 밀도를 지도로 표현하여 지역별 인구 밀도 차이를 파악함

목표 : 막대 그래프의 결과를 인구수로 소팅해 표현하지
방법 : 
- 인구수를 소팅하기

#### 히스토그램(Histogrma)
- 평균 0, 표준편차 1인 가우시안 표준 정규분포 난수 발생 결과 분포 시각화

### 난수 : ramdom
- np.random.normal(loc, scale, size) : 정규 분포에서 난수 생성
    - loc: 평균(mean) 값으로, 생성된 난수들의 평균 지정
    - scale: 표준 편차(standard deviation) 값으로, 생성된 난수들의 표준 편차를 지정
    - size: 생성할 난수의 개수 지정, 생성된 배열의 크기를 결정하는데 사용

```python
import numpy as np

# 랜덤한 데이터 생성
np.random.seed(0)

# 평균 0, 표준편차 1인 가우시안 표준 정규분포 난수 발생
# np.random.rand(n) : 1D matrix
# np.random.rand(m,n) : 2D matrix
data = np.random.randn(1000)

# 히스토그램 생성
plt.hist(data, bins=20, edgecolor='black')

# 그래프 제목과 축 레이블 설정
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 그래프 보여주기
plt.show()
```

### date 난수
- pd.date_range(start, end, freq)
    - start: 생성할 날짜 범위의 시작일을 지정합니다. 문자열 형식인 경우 YYYY-MM-DD 형식을 따라야 합니다.
    - end: 생성할 날짜 범위의 종료일을 지정합니다. 문자열 형식인 경우 YYYY-MM-DD 형식을 따라야 합니다.
    - freq: 날짜 간격(frequency)을 지정

```python
# 날짜 범위 생성 (예시로 2021년 1월부터 2021년 12월까지)
dates = pd.date_range(start='2021-01-01', end='2021-12-31', freq='M')

# 판매량 데이터 생성
np.random.seed(0)
product_a_sales = np.random.normal(loc=100, scale=20, size=len(dates))
product_b_sales = np.random.normal(loc=150, scale=30, size=len(dates))
product_c_sales = np.random.normal(loc=80, scale=15, size=len(dates))

# 데이터프레임 생성
sales_data = pd.DataFrame({
    'Date': dates,
    'Product A': product_a_sales,
    'Product B': product_b_sales,
    'Product C': product_c_sales
})

# 데이터프레임 출력
print(sales_data.head())
```

```python
# 문제 - 제품 A의 판매량 데이터 추출
product_a_sales = sales_data['Product A']
# print(product_a_sales)

# 문제 - 히스토그램 생성, bins=14, edgecolor='black'
plt.hist(product_a_sales, bins=14, edgecolor='black')


# 문제 - 그래프 제목과 x,y 축 레이블 설정
# 그래프 제목과 축 레이블 설정
plt.title('Histogram of Product A Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')


# 그래프 보여주기
plt.show()
```

### 3-3 전체 대비 부분 파악(Comparison to Whole):
- 데이터의 부분이 전체에 대비하여 어떤 비율을 차지하는지 파악하고자 할 때 사용되는 차트 유형
- 원 그래프(Pie Chart), 도넛 그래프(Donut Chart) 등을 사용하여 데이터의 상대적인 비율 시각화

#### 원 그래프(Pie Chart)
- 과일의 크기 비율 시각화

```python
# 데이터
fruits = ['Apple', 'Banana', 'Orange', 'Grape', 'Watermelon']
sales = [30, 40, 25, 15, 20]

# 파이차트 생성
# plt.pie(100%에서 계산할 값, labels=fruits, autopct='%1.소수점자리f%%', startangle=90)
plt.pie(sales, labels=fruits, autopct='%1.1f%%', startangle=120)

# 차트 제목 설정
plt.title('Sales of Fruits')

# 차트 보여주기
plt.show()
```

#### 도넛 그래프(Donut Chart)
- 삼품 카테고리별

```python
# 데이터
categories = ['Category A', 'Category B', 'Category C']
ratios = [30, 40, 20]

# 도넛 그래프 생성
# plt.pie(데이터, labels=레이블, autopct='%비율백분율로표시%', startangle=시작_각도, wedgeprops={'edgecolor': 'white'})
plt.pie(ratios, labels=categories, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'white'})

# 중앙에 흰색 원 생성하여 도넛 모양으로 만들기 
# plt.Circle((0, 0), 0.5, color='white') : 중앙점에서부터 반지름의 0.5 지점까지 흰색으로 채우기
center_circle = plt.Circle((0, 0), 0.5, color='white', edgecolor='white')

# plt.gcf() 함수로 현재 그래프의 Figure 객체를 가져온 뒤 add_artist()로 원을 추가함
fig = plt.gcf()
fig.gca().add_artist(center_circle)

# 그래프 제목 설정
plt.title('Category Ratios')

# 도넛 그래프 보여주기
plt.show()
```

### 3-4 상관관계(Correlation):

### 히트맵(Heatmap):
- fmt='d' 는 정수 형태(integer format)로 숫자를 입력하라는 뜻
- fmt : 값을 표현하는 형식 설정
    - **g** : 일반 숫자
    - **.2f** : 소숫점 이하 2자리
    - **.0f** : 소숫점 이하 자릿수 0 -> 정수 확인
    - **d** : 10진수로 표시
    - **'%':.2f** : 백분율 형식으로 값 표시, 소숫점 이하 2자리까지 표시
- 데이터의 상대적인 크기를 색상으로 시각화하여 나타내는 그래프
- 주로 2차원 데이터를 표현하며, 행과 열로 구성된 격자 형태로 표시
- 데이터의 패턴, 상관관계, 분포 등을 시각적으로 파악할 수 있어 다양한 분야에서 사용됨

```python
# 데이터
data = {
    'Date': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05'],
    'Stock1': [100, 110, 105, 95, 120],
    'Stock2': [80, 85, 90, 95, 100],
    'Stock3': [120, 130, 125, 135, 140],
    'Stock4': [70, 75, 80, 85, 90]
}

# 데이터프레임 생성
df = pd.DataFrame(data)
df
```

```python
# 데이터 인덱스 설정
df.set_index('Date', inplace=True)

# 히트맵 생성
fig, ax = plt.subplots()
im = ax.imshow(df.values, cmap='YlOrRd')

# 축 설정
ax.set_xticks(np.arange(len(df.columns)))
ax.set_yticks(np.arange(len(df.index)))
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.index)

# 텍스트 주석 추가
for i in range(len(df.index)):
    for j in range(len(df.columns)):
        text = ax.text(j, i, df.values[i, j], ha='center', va='center', color='black')

# 컬러바 추가
cbar = ax.figure.colorbar(im, ax=ax)

# 그래프 제목 설정
plt.title('Stock Prices Heatmap')

# 그래프 보여주기
plt.show()
```

### dataframe.corr()
- 피어슨 상관계수(Pearson correlation coefficient) 계산

- 기본적으로 열들 간의 상관계수를 계산하며, 결과는 새로운 데이터프레임으로 반환

- 상관계수 행렬은 정사각형 형태이며, 대각선을 기준으로 대칭

- 선형 상관성을 측정하는 지표, -1에서 1 사이의 값을 가짐

    - 1에 가까울수록 양의 선형 상관관계
    - -1에 가까울수록 음의 선형 상관관계
    - 0에 가까우면 상관관계가 없음        

```python
# HeatMap의 일반적 활용
df_cor = df.corr()

# 히트맵 생성
# cmap 종류 : YlOrRd, viridis, plasma,inferno, magma, cividis, cool, coolwarm, hot, RdYlBu, BuPu
sns.heatmap(df_cor, cmap='viridis', annot=True, fmt='g')

# 그래프 제목 설정
plt.title('Stock Prices Heatmap')

# 그래프 보여주기
plt.show()
```

## 3-5 연결관계(Connection):
- 데이터 간의 연결관계를 시각화하고자 할 때 사용되는 차트 유형
- 네트워크 그래프(Network Graph), 트리맵(Treemap), 국도도(Map of Connections)

### 네트워크 그래프(Network Graph)
- pip install networkx
- 현실 세계에서의 다양한 상호 관계를 시각적으로 나타내는 그래프
- 네트워크 그래프는 노드(Node)와 엣지(Edge)로 이루어져 있으며, 노드는 개별 요소 또는 개체를 나타내고, 엣지는 노드 간의 관계를 나타냄
- 소셜 네트워크, 웹 페이지 간의 링크 구조, 전력 공급망, 교통 네트워크 등 다양한 영역에서 사용

```python
import networkx as nx
# import matplotlib.pyplot as plt

# 노드와 엣지 데이터 생성
nodes = ['A', 'B', 'C', 'D', 'E']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]

# 그래프 생성
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# 그래프 시각화
pos = nx.circular_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.axis('off')
plt.show()
```

### 트리맵(Treemap):
- pip install squarify
- 계층적인 데이터를 사각형의 크기와 색상으로 시각화하는 방법
- 주로 계층적인 구조를 가진 데이터를 표현하고 비교하는 데 사용
- 트리맵은 데이터의 상대적인 크기와 구조를 한눈에 파악할 수 있어 효과적인 정보 전달을 가능

```python
import squarify
# import matplotlib.pyplot as plt

# 데이터와 라벨 생성
data = [400, 300, 200, 100]
labels = ['A', 'B', 'C', 'D']

# 트리맵 생성
plt.figure(figsize=(8, 8))
squarify.plot(sizes=data, label=labels, color=['lightblue', 'lightgreen', 'lightyellow', 'lightpink'])
plt.axis('off')
plt.show()
```

#### 국도도(Map of Connections):
- 노드(Node)와 엣지(Edge)로 이루어져 있으며, 노드는 위치를 나타내고, 엣지는 위치 간의 연결 관계를 나타냄
- 주로 지리적으로 분포된 데이터의 상호 연결성을 시각화하는 데 사용함

```python
# import networkx as nx
# import matplotlib.pyplot as plt

# 노드와 엣지 데이터 생성
nodes = ['A', 'B', 'C', 'D', 'E', 'F']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A'),('E', 'F')]

# 그래프 생성
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# 노드 위치 정보 생성
pos = {'A': (0, 0), 'B': (1, 1), 'C': (2, 0), 'D': (3, 1), 'E': (4, 0), 'F': (5, 1)}

# 그래프 시각화
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```
